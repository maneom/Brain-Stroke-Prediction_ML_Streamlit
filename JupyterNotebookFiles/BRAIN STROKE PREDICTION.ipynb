# ONE HOT ENCODING

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import pickle
```

```python
# CURRENT DATASET HAVE 5 CATEGORICAL COLUMNS. I AM GOING TO USE ONE HOT ENCODER TO CONVERT THEM TO NUMERIC VALUES

transformer = ColumnTransformer(transformers=[
    ('ohe', OneHotEncoder(dtype=int, drop='first'),['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'])
], remainder='passthrough')
X = transformer.fit_transform(df.drop('stroke', axis=1))
y = df['stroke']
```

```python
pickle.dump(transformer,open("transformer.pkl", "wb"))
```

# OVERSAMPLING

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# MODEL CREATION

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_predict = rfc.predict(X_test)
print(classification_report(y_test, y_predict))
```

```python
pickle.dump(rfc,open("RandomForestClassifier.pkl", "wb"))
```
